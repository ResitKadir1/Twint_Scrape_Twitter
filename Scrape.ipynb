## Scrape by username

```python
#Scrape by user name and time intervals with tweets limit
def scrape_by_userName(user_name,time_interval,Limit_of_tweets):
    c =twint.Config()
    c.Username=user_name
    c.Limit =Limit_of_tweets
    c.Since=time_interval
    c.Store_csv= True
    c.Hide_output = True
    c.Output=user_name+".csv"
    c.Format ="Tweet id: {id} | Date: {date} | Time: {time} | Tweet: {tweet} | User ID {user_id} | Reply To {reply_to}"
    twint.run.Search(c) 
    return c
```

```python
scrape_by_userName("AndrewYNg","2019-01-01 00:00:01",10)
```

## Get Followers name 

```python
def get_followers(user_name):
    c =twint.Config()
    c.Username=user_name
    c.Output = user_name+'.csv'
    c.Custom["User"] = ["id","username"]
    twint.run.Followers(c)
    
    return c
```

```python
get_followers("slross58")
```

```python
followers = pd.read_csv("slross58.csv",header=None)
followers
```

```python
def get_following(user_name):
    c =twint.Config()
    c.Username=user_name
    c.Output = user_name+'.csv'
    c.Custom["User"] = ["id","username"]
    twint.run.Following(c)
    
    return c
```

```python
get_following("mohamedzedanels")
```

## Hashtags and contain Word

```python
def scrape_withHashtags(hashtag,start_date,end_date):
    c = twint.Config()
    c.Search = hashtag
    c.Since = start_date
    c.Until = end_date
    c.Hide_output = True
    c.Store_json = True
    c.Output = hashtag+'.json'
    twint.run.Search(c)
  
    return c
```

```python
scrape_withHashtags("blacklivematterUK","2020-06-21","2020-07-22")
```
